```python
# Ensure libraries are installed:
# !pip install h3 matplotlib numpy pyproj

import h3
import math
import random
import matplotlib.pyplot as plt
import numpy as np
import os
import pyproj

# --- Global Configurations ---
STATIC_EPSILON_M_SQ = 1.5
ENABLE_CARPET_TEST = True
NUM_CARPET_POINTS_PER_AXIS = 40
CARPET_PADDING_FACTOR = 0.3

print(f"Running with h3 library version: {h3.__version__}")
try:
    print(f"Pyproj version: {pyproj.__version__}")
    PYPROJ_TRANSFORMER = pyproj.Transformer.from_crs("EPSG:4326", "EPSG:3857", always_xy=True)
except ImportError:
    print("ERROR: pyproj library is not installed. Please install it: pip install pyproj")
    PYPROJ_TRANSFORMER = None
    exit()
except Exception as e:
    print(f"Error initializing pyproj transformer: {e}")
    PYPROJ_TRANSFORMER = None
    exit()

# --- Helper Function: Point-In-Polygon Test ---
def is_point_in_2d_convex_polygon(point_xy_meters, polygon_vertices_xy_meters, 
                                  epsilon_m_sq, 
                                  point_name_for_debug=None, verbose_debug=False):
    px_m, py_m = point_xy_meters
    num_vertices = len(polygon_vertices_xy_meters)
    if num_vertices < 3: return False

    if verbose_debug and point_name_for_debug:
        print(f"    Debugging D_j for {point_name_for_debug} ({px_m:,.1f}m,{py_m:,.1f}m) with static_epsilon={epsilon_m_sq:.3e} m^2:")
    
    for i in range(num_vertices):
        x1_m, y1_m = polygon_vertices_xy_meters[i]
        x2_m, y2_m = polygon_vertices_xy_meters[(i + 1) % num_vertices]
        d_j_m_sq = (x2_m - x1_m) * (py_m - y1_m) - (y2_m - y1_m) * (px_m - x1_m)
        
        if verbose_debug and point_name_for_debug:
            print(f"      Edge {i}-{(i+1)%num_vertices}: D_j = {d_j_m_sq:.3e} m^2")
        if d_j_m_sq < -epsilon_m_sq: 
            if verbose_debug and point_name_for_debug:
                print(f"        ^-- OUTSIDE (D_j < -static_epsilon)")
            return False
            
    if verbose_debug and point_name_for_debug:
        print(f"        INSIDE (all edges passed static epsilon test)")
    return True

# --- Main Analysis Function for a Single H3 Cell ---
def run_analysis_for_cell(h3_address_to_test, current_h3_resolution, current_user_test_point_latlon, 
                          transformer, 
                          static_epsilon_for_test_m_sq, 
                          enable_carpet, num_carpet_pts, carpet_pad_factor):
    print(f"  Starting analysis for H3 cell: {h3_address_to_test} at resolution {current_h3_resolution}")
    boundary_vertices_geo = None # Still needed for UTP processing if it exists
    projected_cell_vertices_wm = [] 
    # center_lat, center_lon = None, None # Not strictly needed if no Folium map
    
    matplotlib_data = {
        'user_test_point_projected_wm': None, 'user_test_point_is_inside': False, 'user_test_point_h3_assigned': None,
        'carpet_points_inside_projected_wm': [], 'carpet_points_outside_projected_wm': []
    }
    # folium_markers = [] # Removed

    h3_specific_error_type = ValueError
    if hasattr(h3, 'error') and hasattr(h3.error, 'H3CellError'): h3_specific_error_type = h3.error.H3CellError
    elif hasattr(h3, 'H3Error'): h3_specific_error_type = h3.H3Error

    try:
        raw_boundary_geo = h3.cell_to_boundary(h3_address_to_test)
        if len(raw_boundary_geo) != 6: 
            print(f"    Cell {h3_address_to_test} is not a hexagon (vertices: {len(raw_boundary_geo)}). Skipping analysis.")
            return

        boundary_vertices_geo = list(raw_boundary_geo) # Keep for UTP context if needed
        # center_lat, center_lon = h3.cell_to_latlng(h3_address_to_test) # Not needed for Folium anymore
        
        for i, (v_lat, v_lon) in enumerate(boundary_vertices_geo):
            wx, wy = transformer.transform(v_lon, v_lat)
            projected_cell_vertices_wm.append((wx, wy))

        if not projected_cell_vertices_wm:
            print("    Warning: Projected cell vertices (Web Mercator) are empty."); return
        
        print(f"    Using static epsilon for test: {static_epsilon_for_test_m_sq} m^2")
        
        print(f"    Projected Web Mercator Vertices (Xm, Ym) for H3 cell {h3_address_to_test} (Order used for test):")
        for i, (vx, vy) in enumerate(projected_cell_vertices_wm):
            print(f"      Vertex {i}: ({vx:,.3f}, {vy:,.3f})")

    except h3_specific_error_type as e: print(f"    Error H3 processing {h3_address_to_test}: {e}"); return
    except Exception as e: print(f"    Error H3/Projection setup {h3_address_to_test}: {e}"); return

    # --- Process User Test Point ---
    if current_user_test_point_latlon:
        utp_lat, utp_lon = current_user_test_point_latlon
        user_point_h3_cell = h3.latlng_to_cell(utp_lat, utp_lon, current_h3_resolution)
        matplotlib_data['user_test_point_h3_assigned'] = user_point_h3_cell
        try:
            utp_wx, utp_wy = transformer.transform(utp_lon, utp_lat)
            matplotlib_data['user_test_point_projected_wm'] = (utp_wx, utp_wy)
            
            temp_is_inside_utp = is_point_in_2d_convex_polygon(
                (utp_wx, utp_wy), projected_cell_vertices_wm, static_epsilon_for_test_m_sq
            )
            verbose_utp_debug = (user_point_h3_cell != h3_address_to_test) or \
                                (temp_is_inside_utp != (user_point_h3_cell == h3_address_to_test))

            is_inside = is_point_in_2d_convex_polygon(
                (utp_wx, utp_wy), projected_cell_vertices_wm, static_epsilon_for_test_m_sq,
                "UserTestPt", verbose_utp_debug
            )
            matplotlib_data['user_test_point_is_inside'] = is_inside
            
            print(f"    User Test Point (Yellow Star) Lat/Lon: ({utp_lat:.6f}, {utp_lon:.6f})")
            print(f"    User Test Point (Yellow Star) Projected Web Mercator: (X: {utp_wx:,.3f} m, Y: {utp_wy:,.3f} m)")
            print(f"    User Test Point (Yellow Star) H3 Assigned: {user_point_h3_cell}, Geo-Test Result: {'INSIDE' if is_inside else 'OUTSIDE'} vs {h3_address_to_test}")

            # user_popup logic removed as folium_markers is removed
        except Exception as e_proj_utp: print(f"    Error processing User Test Point: {e_proj_utp}")

    # --- Carpet Test ---
    if enable_carpet and projected_cell_vertices_wm:
        min_x_wm=min(v[0] for v in projected_cell_vertices_wm); max_x_wm=max(v[0] for v in projected_cell_vertices_wm)
        min_y_wm=min(v[1] for v in projected_cell_vertices_wm); max_y_wm=max(v[1] for v in projected_cell_vertices_wm)
        width_wm = max_x_wm - min_x_wm; height_wm = max_y_wm - min_y_wm
        if abs(width_wm) < 1.0: width_wm = abs(max_x_wm * 0.1) if abs(max_x_wm) > 1.0 else 10.0
        if abs(height_wm) < 1.0: height_wm = abs(max_y_wm * 0.1) if abs(max_y_wm) > 1.0 else 10.0
        if width_wm == 0: width_wm = 10.0; 
        if height_wm == 0: height_wm = 10.0
        grid_min_x_wm=min_x_wm-width_wm*carpet_pad_factor; grid_max_x_wm=max_x_wm+width_wm*carpet_pad_factor
        grid_min_y_wm=min_y_wm-height_wm*carpet_pad_factor; grid_max_y_wm=max_y_wm+height_wm*carpet_pad_factor
        carpet_x_coords_wm = np.linspace(grid_min_x_wm, grid_max_x_wm, num_carpet_pts)
        carpet_y_coords_wm = np.linspace(grid_min_y_wm, grid_max_y_wm, num_carpet_pts)
        for cy_wm in carpet_y_coords_wm:
            for cx_wm in carpet_x_coords_wm:
                is_inside = is_point_in_2d_convex_polygon(
                    (cx_wm, cy_wm), projected_cell_vertices_wm, static_epsilon_for_test_m_sq 
                )
                if is_inside: matplotlib_data['carpet_points_inside_projected_wm'].append((cx_wm, cy_wm))
                else: matplotlib_data['carpet_points_outside_projected_wm'].append((cx_wm, cy_wm))
    
    # --- Matplotlib Plot ---
    if projected_cell_vertices_wm: 
        fig, ax = plt.subplots(figsize=(10,10))
        cell_xs_wm=[v[0] for v in projected_cell_vertices_wm]+[projected_cell_vertices_wm[0][0]]
        cell_ys_wm=[v[1] for v in projected_cell_vertices_wm]+[projected_cell_vertices_wm[0][1]]
        ax.plot(cell_xs_wm, cell_ys_wm, 'b-', linewidth=2, label=f'H3 Cell Boundary')
        
        for i, (vx, vy) in enumerate(projected_cell_vertices_wm):
            ax.plot(vx, vy, 'o', color='blue', markersize=7, mfc='lightblue', zorder=3) 
            ax.text(vx, vy, f' V{i}', color='black', ha='left', va='bottom', fontsize=9, zorder=4,
                    bbox=dict(boxstyle="round,pad=0.1", fc="lightblue", ec="blue", alpha=0.7))
        
        if enable_carpet:
            if matplotlib_data['carpet_points_inside_projected_wm']:
                xs=[p[0] for p in matplotlib_data['carpet_points_inside_projected_wm']]
                ys=[p[1] for p in matplotlib_data['carpet_points_inside_projected_wm']]
                ax.scatter(xs,ys,c='lightgreen',marker='o',s=8,label='Carpet Points: Inside Test',alpha=0.7,zorder=1)
            if matplotlib_data['carpet_points_outside_projected_wm']:
                xs=[p[0] for p in matplotlib_data['carpet_points_outside_projected_wm']]
                ys=[p[1] for p in matplotlib_data['carpet_points_outside_projected_wm']]
                ax.scatter(xs,ys,c='lightcoral',marker='x',s=8,label='Carpet Points: Outside Test',alpha=0.7,zorder=1)
        if matplotlib_data['user_test_point_projected_wm']:
            up_proj_wm=matplotlib_data['user_test_point_projected_wm']
            up_is_inside=matplotlib_data['user_test_point_is_inside']
            ax.scatter(up_proj_wm[0],up_proj_wm[1],edgecolor='black',facecolor='yellow',marker='*',s=200,label='User Test Point',zorder=5)
            x_range = ax.get_xlim()[1]-ax.get_xlim()[0]; y_range = ax.get_ylim()[1]-ax.get_ylim()[0]
            text_offset_x = x_range*0.02 ; text_offset_y = y_range*0.02
            ax.text(up_proj_wm[0]+text_offset_x,up_proj_wm[1]+text_offset_y,f" UTP\n GeoTest:{'IN' if up_is_inside else 'OUT'}",
                     fontsize=8,ha='left',va='bottom',color='black',zorder=5,bbox=dict(boxstyle="round,pad=0.2",fc="yellow",ec="black",alpha=0.7))
        
        ax.set_title(f'H3 Cell: {h3_address_to_test} (Res {current_h3_resolution}) - Web Mercator\nStatic Epsilon for Test: {static_epsilon_for_test_m_sq} m^2')
        ax.set_xlabel('Web Mercator X (meters)'); ax.set_ylabel('Web Mercator Y (meters)')
        ax.legend(loc='best',fontsize='small'); ax.grid(True); ax.axis('equal')
        print(f"    Displaying Matplotlib plot for res {current_h3_resolution}, cell {h3_address_to_test}...")
        plt.show(); plt.close(fig)

    print(f"  Finished analysis for H3 cell: {h3_address_to_test}")

# --- Main Execution Block ---
if __name__ == "__main__":
    if PYPROJ_TRANSFORMER is None: print("Pyproj transformer not initialized. Cannot proceed."); exit()

    print("\n--- Global Web Mercator Coordinate Range Test ---")
    print("Theoretical Web Mercator X/Y range is approx. +/- 20,037,508 meters.")
    print("Displaying sample coordinates with 3 decimal places (millimeter indication):")
    global_test_points = [
        ("Equator, Prime Meridian",0.0,0.0),("Equator, Intl Dateline",0.0,180.0),("Equator, -Intl Dateline",0.0,-180.0),
        ("Web Mercator Max Lat",85.05113,0.0),("Web Mercator Min Lat",-85.05113,0.0),("Mid Lat N, W",45.0,-90.0),
        ("Mid Lat S, E",-45.0,90.0),("High Lat N (80N)",80.0,45.0),("High Lat S (-80S)",-80.0,-45.0),]
    for name, lat_deg, lon_deg in global_test_points:
        try:
            wm_x, wm_y = PYPROJ_TRANSFORMER.transform(lon_deg, lat_deg) 
            print(f"  {name:<28s} Lat/Lon ({lat_deg:6.2f}, {lon_deg:7.2f}) -> Web Mercator (X: {wm_x:15,.3f} m, Y: {wm_y:15,.3f} m)")
        except Exception as e_proj_global: print(f"  Error projecting {name} Lat/Lon ({lat_deg:.1f}, {lon_deg:.1f}): {e_proj_global}")
    print("--- End of Global Coordinate Range Test ---\n")

    geo_points_to_try = [(0.0,0.0),(0.0,90.0),(0.0,-90.0),(0.0,45.0),(0.0,-45.0),(0.0,135.0),(0.0,-135.0),
                         (45.0,0.0),(45.0,90.0),(45.0,-90.0),(45.0,180.0),(-45.0,0.0),(-45.0,90.0),
                         (-45.0,-90.0),(-45.0,180.0),(70.0,0.0),(-70.0,0.0),(20.0,20.0),(-20.0,-20.0)]

    for res in range(16): 
        print(f"\n--- Processing Resolution {res} ---")
        current_cell_to_test = None; user_test_point_for_run = None
        selected_from_geopoint_info = "N/A"
        for attempt_idx, (lat, lon) in enumerate(geo_points_to_try):
            try:
                candidate_cell = h3.latlng_to_cell(lat, lon, res)
                boundary = h3.cell_to_boundary(candidate_cell) 
                if len(boundary) == 6: 
                    current_cell_to_test = candidate_cell
                    selected_from_geopoint_info = f"(Lat:{lat:.2f}, Lon:{lon:.2f})"
                    break 
            except Exception as e_cell_select: continue
        
        if current_cell_to_test is None:
            print(f"  Could not find a hexagonal cell for res {res} from predefined points. Skipping.")
            continue
        print(f"  Selected cell for res {res}: {current_cell_to_test} (Hexagon found near {selected_from_geopoint_info})")
        try:
            temp_boundary = h3.cell_to_boundary(current_cell_to_test)
            if temp_boundary: user_test_point_for_run = temp_boundary[0]
        except Exception as e_utp_boundary: print(f"  Warning: Could not get boundary for {current_cell_to_test} to set UTP: {e_utp_boundary}")
        
        run_analysis_for_cell(
            current_cell_to_test, res, user_test_point_for_run,
            PYPROJ_TRANSFORMER, 
            STATIC_EPSILON_M_SQ, 
            ENABLE_CARPET_TEST, NUM_CARPET_POINTS_PER_AXIS, CARPET_PADDING_FACTOR,
        )
    
    print("\n--- All resolutions processed ---")
```
